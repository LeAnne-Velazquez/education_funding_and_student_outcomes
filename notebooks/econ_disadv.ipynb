```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import numpy as np
from scipy.stats import pearsonr
```

#### GRADE 4 MATH NAEP

```python
gr_4_math_disadv = pd.read_excel('../data/gr_4_math_disadv_naep_raw.xls', engine ="xlrd")
```

```python
gr_4_math_disadv = gr_4_math_disadv.iloc[8:1724]
gr_4_math_disadv
```

```python
gr_4_math_disadv.columns = ['Year', 'State', 'Econ_Disadv_Status', 'Avg_Scale_Score']

gr_4_math_disadv = gr_4_math_disadv[gr_4_math_disadv['Econ_Disadv_Status'] != 'Information not available']
```

```python
gr_4_math_disadv['Avg_Scale_Score'] = gr_4_math_disadv['Avg_Scale_Score'].replace('‡', np.nan)
```

```python

gr_4_math_disadv['Year'] = gr_4_math_disadv['Year'].astype(int)
gr_4_math_disadv['Avg_Scale_Score'] = gr_4_math_disadv['Avg_Scale_Score'].astype(float)


gr_4_math_disadv['Avg_Scale_Score'] = gr_4_math_disadv['Avg_Scale_Score'].round(1)


gr_4_math_disadv = gr_4_math_disadv.reset_index(drop=True)
```

```python
gr_4_math_disadv
```

```python
gr_4_math_disadv.to_csv("../data/gr_4_math_disadv_naep_clean.csv", index=False)
```

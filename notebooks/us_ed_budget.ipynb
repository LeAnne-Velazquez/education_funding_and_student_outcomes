```python
dfs = [doe_2000, doe_2001_2008, doe_2009, doe_2010, doe_2011, doe_2012, doe_2013, doe_2014, doe_2015, doe_2016, doe_2017, doe_2018, doe_2019, doe_2020, doe_2021, doe_2022, doe_2023, doe_2024]

ed_merged_wide = pd.concat(
    [df.set_index("State or Other Area") for df in dfs], 
    axis=1
).reset_index()
```

```python
ed_merged_wide
```

```python
ed_merged_long = ed_merged_wide.melt(
    id_vars="State or Other Area",   # keep this column as is
    var_name="Year",                 # new column name for former wide column headers
    value_name="Value"               # new column name for the numbers
)
```

```python
ed_merged_long
```

```python
ed_merged_long.rename(columns={"State or Other Area": "State"}, inplace=True)
```

```python
ed_merged_long
```

```python
ed_merged_long["Year"] = ed_merged_long["Year"].astype(int)

inflation_factors = {
    2000: 1.83, 2001: 1.76, 2002: 1.74, 2003: 1.70, 
    2004: 1.67, 2005: 1.62, 2006: 1.56, 2007: 1.52,
    2008: 1.46, 2009: 1.46, 2010: 1.42, 2011: 1.40,
    2012: 1.36, 2013: 1.34, 2014: 1.32, 2015: 1.32,
    2016: 1.30, 2017: 1.27, 2018: 1.24, 2019: 1.23,
    2020: 1.20, 2021: 1.18, 2022: 1.10, 2023: 1.03,
    2024: 1.00
}

ed_merged_long["Inflation_Factor"] = ed_merged_long["Year"].map(inflation_factors)

ed_merged_long["Value_adj"] = ed_merged_long["Value"] * ed_merged_long["Inflation_Factor"]
```

```python
ed_merged_long.head(50)
```

```python
ed_merged_long.to_csv("../data/ed_merged_long.csv", index=False)
```

```python
ed_merged_filtered = ed_merged_long[
    (ed_merged_long["Year"] >= 2000) & 
    (ed_merged_long["Year"] <= 2021)
]


state_avg = (
    ed_merged_filtered.groupby("State")["Value_adj"]
    .mean()
    .round(1)
    .reset_index()
)


state_avg_sorted = state_avg.sort_values(by="Value_adj", ascending=False).reset_index(drop=True)
```

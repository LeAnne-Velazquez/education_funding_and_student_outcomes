#### NAEP 4 Reading Scores

```python
grade_4_reading = pd.read_excel('../data/naep_4_reading.xls', engine ="xlrd")
```

```python
grade_4_reading.info()
```

```python
grade_4_reading = grade_4_reading.iloc[8:620]
```

```python
grade_4_reading = grade_4_reading.rename(columns={
    grade_4_reading.columns[0]: "Year",
    "Unnamed: 1": "State",
    "Unnamed: 3": "Avg_Scale_Score"
})
```

```python
grade_4_reading = grade_4_reading.drop(columns=["Unnamed: 2"])
```

```python
grade_4_reading["Avg_Scale_Score"] = grade_4_reading["Avg_Scale_Score"].replace(["—", "‡"], np.nan)
```

```python
grade_4_reading["Avg_Scale_Score"] = grade_4_reading["Avg_Scale_Score"].astype(float)
grade_4_reading["Year"] = grade_4_reading["Year"].astype(int)
grade_4_reading["Avg_Scale_Score"] = grade_4_reading["Avg_Scale_Score"].round(1)
```

```python
grade_4_reading.tail(50)
```

```python
grade_4_reading.to_csv("../data/grade_4_reading.csv", index=False)
```

##### grade 4 reading sorted & filtered 00-24 

```python
filtered_df = grade_4_reading[(grade_4_reading['Year'] >= 2000) & (grade_4_reading['Year'] <= 2024)]


avg_scores = (
    filtered_df.groupby('State', as_index=False)['Avg_Scale_Score']
    .mean()
)

avg_scores['Avg_Scale_Score'] = avg_scores['Avg_Scale_Score'].round(1)

avg_4_reading = avg_scores.sort_values(by='Avg_Scale_Score', ascending=False).reset_index(drop=True)
```

```python
avg_4_reading
```

```python
avg_4_reading.to_csv("../data/avg_4_reading.csv", index=False)
```
